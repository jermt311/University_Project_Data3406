### Conversion to data time.

```python
#user_1['Start'] = pd.to_datetime(user_1['Start'], format='%d-%b-%Y %H:%M')
#user_1['Finish'] = pd.to_datetime(user_1['Finish'], format='%d-%b-%Y %H:%M')
```

```python
#db['Start'] = pd.to_datetime(db['Start'], format='%d-%b-%Y %H:%M')
db['Finish'] = pd.to_datetime(db['Finish'], format='%d-%b-%Y %H:%M')
```

Converting Finish column to date time

```python
# Add columns with year, month, and weekday name
db['Year'] = db.index.year
db['Month'] = db.index.month
#db['Weekday Name'] = db.index.weekday_name
db['Weekday Name'] = db.index.day_name()
```

Creation of new columns Year, Month and Weekday Name. For easier visualisation

## Data Visualisation

### Trend in yearly Step count of User

```python
# plot data
fig, ax = plt.subplots(figsize=(15,7))
# use unstack()
db.groupby(['Year'])['Steps (count)'].sum().plot(ax=ax)
plt.title("Yearly Step count")
plt.xlabel("Year")
plt.ylabel("Step Count")
plt.xticks(rotation=45)
```

We can observe that the User slowly started increasing their step count as the years went on. At a quick glance, we can observe that as the years go by, this user has greatly increased his yearly step count

### Which is the day that has the most steps taken by User 1?

```python
# plot data
fig, ax = plt.subplots(figsize=(15,7))
# use unstack()
db.groupby(['Weekday Name','Year'])['Steps (count)'].sum().unstack().plot(ax=ax)
plt.title("Step count by Day sorted by Year")
plt.xlabel("Year")
plt.ylabel("Step Count")
plt.xticks(rotation=45)
```

We can observe that the user walks the most on Thursdays and Fridays compared to the other days over the years.

### User's Monthly Step Count. 

```python
# plot data
fig, ax = plt.subplots(figsize=(15,7))
# use unstack()
db.groupby(['Month'])['Steps (count)'].sum().plot(ax=ax)
plt.title("Monthly Step count")
plt.xlabel("Year")
plt.ylabel("Step Count")
```

We can observe that the user walks the most in March and the least in November

### The User's Yearly Step count Grouped by Day

```python
# plot data
fig, ax = plt.subplots(figsize=(15,7))
# use unstack()
db.groupby(['Year','Weekday Name'])['Steps (count)'].sum().unstack().plot(ax=ax)
plt.title("Yearly Step count grouped by Day")
plt.xlabel("Year")
plt.ylabel("Step Count")
```

It appeats that the user walks the least on Saturdays compared to the other days as the years go by.
